```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Build the mesh
x_dim = 10
y_dim = 10
x = np.linspace(-x_dim, x_dim, 500)
y = np.linspace(-y_dim, y_dim, 500)
xx,yy = np.meshgrid(x,y)
```

```python
#Define our point source
wavelength = 1
wavenumber = 2*np.pi
amplitude = 1
z = amplitude * np.exp(1j * wavenumber * np.sqrt(xx**2 + yy**2))
```

```python
fig,ax = plt.subplots(1,1, figsize=(10,10))
ax.imshow(np.real(z))
ax.axvline(x=100, ymin=0, ymax=10, lw=5, c='r')
```

```python
#Projection on the line
xx_ = -1
z = amplitude * np.exp(1j * wavenumber * np.sqrt(xx_**2 + y**2))

plt.plot(y,np.real(z))
```
